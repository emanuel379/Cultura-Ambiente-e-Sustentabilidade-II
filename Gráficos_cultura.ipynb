```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dados de arborização nos bairros da cidade de São Carlos( Meta 30%)

dados = [ 29.6, 79.55, 15.39, 9.81, 11.68, 12.31, 27.51, 10.13, 10.13, 8.15, 37.31, 13.94, 14.45, 21.45, 13.62, 13.92, 7.23, 7.31,
         8.54, 7.69, 11.2, 12.76, 19.06, 16.63, 14.48, 18.3, 5.77, 8.08, 8.91, 13.61, 8.31, 6.28, 9.33, 7.42, 10.19,
         11.05, 10.76, 18.98, 13.52, 7.22, 7.65, 17.44, 51.17, 4.88, 9.01]

dados = pd.DataFrame({'Arborização': dados})
dados
```

```python
dados.mean()
```

```python
dados.median()
```

```python
# Analizando o boxplot com a dispersão dos dados

plt.figure(figsize=(10, 8))
sns.boxplot(data = dados, showmeans=True, meanprops={"marker": "*", "markeredgecolor": "black", "markersize": "10"}).set(title="Análise da cobertura vegetal em São Carlos")
plt.ylabel("áreas verdes")
plt.show()
```

```python
# Observando um gráfico de linhas

dados.plot(label = 'Cobertura arbórea')
plt.axhline(y = 30, color = 'red')
plt.axhline(y = 21.48, color = 'blue', linestyle ="--")
plt.axhline(y = 11.2, color = 'purple', linestyle =":")
```
